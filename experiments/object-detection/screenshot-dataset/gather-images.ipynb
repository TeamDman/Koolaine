```python
from pathlib import Path
from PIL import Image
from tqdm import tqdm
import shutil
```

```python
# Parameters
width, height = 1920, 1080
source_folder = Path("C:/Users/TeamD/OneDrive/Pictures/Screenshots/ShareX Screenshots")
destination_folder = Path("./filtered-screenshots/")
```

```python
# Create the destination folder if it doesn't exist
destination_folder.mkdir(parents=True, exist_ok=True)
```

```python
%pip install icecream
```

```python
to_check = list(source_folder.glob("2023-*/*.png"))
from icecream import ic
ic(to_check[:10])
```

```python
with Image.open(to_check[0]) as img:
    ic(img.size)
    ic(img.size == (width, height))
```

```python
%pip install loguru
```

```python
# Iterate through each screenshot in the source folder
# only do 2023, older stuff is less relevant and might have been evacuated by onedrive
from loguru import logger
for screenshot in tqdm(to_check):
    try:
        if screenshot.stat().st_size > 0:  # Check if file size is greater than 0
            with Image.open(screenshot) as img:
                if img.size == (width, height):
                    # Copy the file to the destination folder
                    logger.debug(f"Copying {screenshot} to {destination_folder / screenshot.name}")
                    shutil.copy(screenshot, destination_folder / screenshot.name)
    except Exception as e:
        print(f"An error occurred while processing {screenshot}: {e}")

print(f"Copied all {width}x{height} screenshots to {destination_folder}")
```
